```python
# import some necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt            
%matplotlib inline     
sns.set(color_codes=True)
```

```python
#import the data file
data = pd.read_csv(r"C:\Users\Aarya Y Deshpande\Downloads\data.csv\data.csv")
#display the top 7 rows
data.head(7)
```

```python
#display the bottom 7 rows 
data.tail(7)
```

```python
#check the data types
data.dtypes
```

```python
#drop the columns which are irrelevent to us
data=data.drop(['Engine Fuel Type', 'Market Category', 'Vehicle Style', 'Popularity', 'Number of Doors', 'Vehicle Size'], axis=1)
data.head(7)
```

```python
#rename the columns for an easy read 
data=data.rename(columns={"Engine HP": "Horse Power", "Engine Cylinders": "Cylinders" ,"Transmission Type": "Transmission","Driven_Wheels": "Drive Mode", "MSRP":"Price"})
data.head(7)
```

```python
#remove the duplicate rows 
data.shape
duplicate_rows=data[data.duplicated()]
print("The number of duplicate rows is:",duplicate_rows.shape)
```

```python
data=data.drop_duplicates()
data.count()
data.head(7)
```

```python
#data about null values
print(data.isnull().sum())
```

```python
#drop the missing/null values
data=data.dropna()
data.count()
```

```python
print(data.isnull().sum())
```

```python
# find unique values and their counts in each column in data using value counts fn
for i in data.columns:
    print ("--------------- %s ----------------" % i)
    print(data[i].value_counts())
```

```python
#Make a Heatmap
data1 = data.loc[:, data.dtypes != object]
corr = data1.corr()
corr
```

```python
plt.figure(figsize = (12,8))
sns.heatmap(corr, cmap = 'BrBG', annot = True)
plt.show()
#Inference: There is a strong positive correlation between horse power & cylinders, and highway mpg & city mpg.
#There is a strong negative correlation between price & city mpg, and price & highway mpg. 
```

```python
sns.histplot(data['Cylinders'])
```

```python
sns.barplot(x=data['Price'])
```

```python
fig, ax = plt.subplots(figsize=(10,6))
ax.scatter(x=data['Horse Power'], y=data['Price'])
ax.set_xlabel('Horse Power')
ax.set_ylabel('Price')
plt.show()
#Inference: There is a positive correlation between the values of horse power and price but the relationship is not exactly linear. 
#Horsepower is a significant factor in determining price
```

```python
#Plot features against each other 
data.Make.value_counts().nlargest(40).plot(kind='bar', figsize=(10,5))
plt.title("Number of cars by make")
plt.ylabel('Number of cars')
plt.xlabel('Make');
#Inference: The maximum no. of cars are of the 'Chevrolet' mak and the mimimum no. of cars of the 'Lotus' make.
```
